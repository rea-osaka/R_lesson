## 現在のディレクトリ

コンピュータ上で作業する時、
自分が今どのディレクトリで作業しているのかを把握する事はいつでも重要な事です。

自分が今作業しているディレクトリの事は、「カレントディレクトリ」、「ワーキングディレクトリ」、「現在のディレクトリ」等と呼ばれます。

Ｒには、次の様なディレクトリ作業をする基礎的な関数があります。

|関数|説明|
|:--|:--|
| getwd() | 現在のディレクトリを表示|
| setwd() | 現在のディレクトリをセット|
| dir() | 現在のディレクトリの中身を表示|

（実習）
 - 現在のディレクトリを表示して下さい
 - 現在のディレクトリにあるファイルのリストを表示してください


## ワーキングディレクトリとは

さて、githubのrea-osakaで紹介されているＲのスクリプトは、
コンソールからそのスクリプトをsource()関数で読み込んで使います。

まずは、githubのリポジトリをダウンロードしてみましょう。
[https://github.com/rea-osaka/rea-osaka-R.module](https://github.com/rea-osaka/rea-osaka-R.module)
から、リポジトリをzipでダウンロードして下さい。
ダウンロード出来たらデスクトップに解凍してください。

解凍されたフォルダの中に幾つかのファイルに交じって、
「REanalysis.R」というファイルがありますが、
これをsource()関数でコンソールで読み込むことで、
色々な便利な関数が使えるようになります。

この読み込むという作業を行うために、
その作業の場所を**現在のディレクトリとして設定**します。

そして、この現在ディレクトリの設定のための関数がsetwd()関数です。
この関数は、例えば次の様に、
カッコの中に引数に作業ディレクトリとなる場所を指定して使います。

```
setwd('c:/Users/ms201/Desktop/rea-osaka-R.module-master/')
```

ここで、ディレクトリの文字列を書き込む際、タブキーを使うことで補完操作ができます。


（実習）
 - 展開したディレクトリを現在ディレクトリに指定してください

## データフレームの構造をみる

データフレーム型のデータはエクセルシートの様に、列と行からなる表のような構造をしています。

データフレームを観察するための基本的な関数は次の通り

|関数|効果|
|:--|:--|
|str()|構造をみる|
|nrow()|何行か|
|ncol()|何列か|
|names()|列名|

Rでは、大きなデータを扱っても大丈夫なようになっています。

例えば、「osaka」データには、30000件以上のデータを含んでいるため、単純にデータの一覧を表示したとすると、30000行以上が表示されることになりますが、通常、これらのデータをコンピュータ上のエクセル等で表示するための処理時間は非常に大きかったり、そもそも表示させるとエラーになったりします。また、もし実際に表示したとしても、その著しく大量のデータというものは、人がざっと見通せる量ではありません。

しかし、Ｒでは、これらの大量データについても見通しが良くなる表示を簡単に行ってくれます。

この、**データを見る**時に役立つ関数の一つが**str()**関数です。
strはstructureの略で「構造」を表しています。
str()関数の出力は、１行目に、データ件数を表す行数、項目数を表す列数が表示されます。
次の行からは、項目名が示され、続いてその項目に含まれるデータの型と、具体的な値の概要が表示されます。

Rで扱うデータは、１行が１件のデータを表しています。
即ち、何行あるかわかれば、何件のデータがあるかが分かります。
列はそれぞれ、そのデータに含まれている項目を示しています。
Rで扱うデータフレームの基本的な構造について、縦横が逆転することはなく、
行と列の意味はこのように決まっています。


（実習）
 - osakaデータがどんな構造か概要を表示してください
 - osakaデータは、何行のデータがありますか
 - osakaデータには、項目が何列ありますか
 - osakaデータには、どんな項目がありますか

## ヒストグラム

ヒストグラムというグラフはデータの分布の具合を教えてくれます。
ヒストグラムはhist()関数に、集合データを渡すことでグラフ表示してくれます。

例えば、枚方市の取引総額データを渡して表示する場合次のようになります。

```
hist(hirakata$取引総額)
```

(実習)
 - 枚方市の取引総額のヒストグラムを表示してください


このヒストグラムは横軸に総額の価格帯、縦軸にその価格帯にある事例数を表していています。

「0.0e+00」表記は、大きな数を表す表記の仕方で、
eの左の数字に10のeの右の数字乗した数字を表します。
例えば、「1.5e+08」は1.5×10^8を表します。
簡単に言えば、eの右にある数字だけ0を並べた桁の数字なので、
08なら0が８個で億、07なら0が7個で千万です。

上記ヒストグラムからは、総額1億円前後以上の事例が存在しますが、
その頻度が極端に小さいことが分かります。


## subset()関数を活用する

## ggplot

Rで奇麗なグラフを書くために、また、グラフを書くための文法が統一されているggplotパッケージが人気です

先ほど、頻度を視覚化するためのヒストグラムを紹介しましたが、
これを個別（離散的）の頻度ではなく密度としてグラフを書くことで、滑らかな曲線のグラフとして表示されます。

ggplotによる描画の命令は以下のようになります。

```
ggplot(hirakata) + geom_density(aes(x=取引総額))
```

まずは、ggplotの引数にデータフレームデータを渡します。次に、描きたいグラフに対応する関数を呼び出し、その引数には、そこに描く抽象図形データを渡します。このグラフ描画関数に渡す引数は、具体的にはaes関数の戻り値であり、このあたりがggplotの独特の文法になっています。

抽象図形データが何から構成されるかは、aes関数に渡す引数により決定されますが、ggplotに渡したデータフレームのうち注目したい列名を渡すことになります。

（実習）
 - 実際に枚方市の取引総額に関する密度曲線を描画してみましょう


## 簡単に比較できるggplot

ggplotの楽しい処の一つは、色を活用した、視覚的にわかりやすい比較が簡単にできるところです。

例えば、osakaデータの中から、京阪方面の郊外地域でである「守口市」、「門真市」「枚方市」、「寝屋川市」、「交野市」の５市についての取引事例データを用意し「cities_data」とします。

cities_dataには５市分のデータが混在していますが、ggplotを使うとこれらを自動的に整理して表示してくれます。

```
ggplot(mydata) + geom_density(aes(x=取引総額, color=市名))
```

ggplotの渡す元データは、cities_dataです。

geom_desity()関数に渡すaes()関数の引数に、「color=市名」となっているところだけが、先の命令と異なっています。
そして、まさにこれだけの命令で、市名要素毎に色を付けて、グラフを書き分けてくれます。


（実習）
 - 上記５市の取引データを含む５つのデータフレームを作成してください
 - ５市分のデータフレームを結合したデータフレーム「cities_data」を作成してください。データフレームの結合はrbind()関数に結合したいデータフレームを渡すことで行えます。
 - 「cities_data」を使って、色分けされた取引総額の密度曲線を描画してください
