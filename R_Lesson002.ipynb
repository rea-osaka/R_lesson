## データを抜き出す仕組み

さて、このベクトル演算。
既にsubset()関数によるデータの抜き出しで、使われていました。

```
nayagawa <- subset(osaka, osaka$市名 == "寝屋川市")
```
という命令です。

subset関数の第１引数は、「osaka」であり、データフレームです。

そして、第２引数の部分は実は、「osaka」の行数と同じだけの数の論理値（TRUEかFALSEのどちらかの値）が３万余り入った、ベクトルです。

### subset()関数の第２引数

まず、osaka$市名は、市名列を抜き出したものであり、市の名前を値とする３万個余りのベクトルです。次のようなprint()関数を使うことで、確かめることが出来ます。

```
print(osaka$市名)
```

このベクトルに対し、「==」を使って、条件ベクトル演算を行うことで、論理値のベクトルを得ています。

やっていることは、先の実習と同じことです。

（実習）
  - osaka$市名 == "寝屋川市"　の結果として得られている論理値ベクトルをprint()関数で確認してみよう








実際に行うと、ほとんどが「FALSE」の値ですが、ざっとスクロールすると23341番目前後から「TRUE」の値を見ることが出来ます。

subset()関数の第２引数は、別に条件式を取るというルールがあるわけではなく、論理値ベクトルを与えることになっています。そして、第１引数で与えられたデータフレームについて、第２引数の論理ベクトルがTRUEになっている**行**を抜き出す働きを持っているのです。

つまり、上でみた、23341番目前後のTRUEに対応した、osakaの23341行目前後が単純に抜き出されているのです。

## データを抜き出す演算子

「$」マークを使うと列が抜き出せる事は紹介しました。

この他にデータフレームから抜き出すための演算子の一つが「[]」です。

データフレームをイメージする時、常にエクセルシートを想像すると分かりやすいですが、
osakaデータの２行目、４列の値を抜き出すためには

```
osaka[2,4]
```

つまり、演算子[]のルールは次のようになっています。

<p>
<div style="text-align: center;">
データフレーム[抜き出したい行の番号, 抜き出したい列の番号]
</div>
<p>

そして、複数の行や列を選択したい場合には

```
osaka[c(2,3,4), c(1,2)]
```
という風にベクトルで指定できます。この場合、2,3,4の３行について、1列目と２列目の項目からなるデータフレームが返されます

次に、全部の列や全部の行を表現したい場合、何も書き込まないことで実現します。
例えば、1行目について全部の列を残したい場合

```
osaka[1, ]
```
と表現されます。

列の指定には、列の番号だけでなく、列名を使うことが出来ます。

また、通常[]演算子での戻り値は、もとのデータフレーム型が戻りますが、
列について１列のみを選択した場合、戻り値は、ベクトルになります。

```
osaka[c(1,2),"取引総額"]
```

（実習）
 - []演算子を使って、色々なデータを取り出してみましょう
